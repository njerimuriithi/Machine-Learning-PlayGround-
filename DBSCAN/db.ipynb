```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mall_customers.csv')
df.head()
```

```python
x= df[['Annual Income (k$)','Spending Score (1-100)']].values
x
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled =scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.5,min_samples=5)
labels = dbscan.fit_predict(x_scaled)
labels
```

```python
np.unique(labels) #-1 means an outlier
```

```python
df['Cluster']=labels
df
```

```python
df.describe()
```

```python
df1=df[df['Cluster']==-1]
df2=df[df['Cluster']==0]
df3=df[df['Cluster']==1]
```

```python
plt.scatter(df1['Annual Income (k$)'],df1['Spending Score (1-100)'],color='black',label='outlier')
plt.scatter(df2['Annual Income (k$)'],df2['Spending Score (1-100)'],color='orange',label='cluster 0')
plt.scatter(df3['Annual Income (k$)'],df3['Spending Score (1-100)'],color='red',label='cluster 1')
plt.legend()
```
