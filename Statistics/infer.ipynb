Inferential is giving an estimate based on the sample from a large population

```python
from math import sqrt
```

```python
sample_size =40
satisfied =32
```

```python
sample_proportion = satisfied/sample_size
z =1.96#95% Confidence leve
```

```python
sample_proportion
```

```python
standard_error = sqrt((sample_proportion*(1- sample_proportion))
/sample_size)
```

```python
standard_error
```

```python
lower = sample_proportion- z* standard_error
upper = sample_proportion+ z* standard_error
```
